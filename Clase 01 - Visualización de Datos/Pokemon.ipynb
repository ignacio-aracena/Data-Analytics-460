```python
plt.figure(figsize=(11,6))
orden = df.groupby('Type 1')['Attack'].median().sort_values().index
colores = sns.color_palette("crest", n_colors=len(orden)) # los colores no significan nada
color_dict = dict(zip(orden, colores))
sns.boxplot(y='Type 1', x='Attack', data=df, order=orden, 
            palette=color_dict, hue='Type 1', dodge=False)
plt.ylabel('Type 1')
plt.xlabel('Attack')
plt.title('Boxplot de Attack')
plt.legend([],[], frameon=False)
plt.show()
```

```python
top10_attack = df.sort_values('Attack', ascending=False).head(10)
plt.figure(figsize=(12,7))
colores = sns.color_palette("flare", n_colors=10)
bars = plt.barh(top10_attack['Name'][::-1], top10_attack['Attack'][::-1], color=colores)
plt.xlabel('Attack')
plt.title('Top 10 Pokemon con mas Attack')
for bar, atk in zip(bars, top10_attack['Attack'][::-1]):
    plt.text(bar.get_width() + 1, bar.get_y() + bar.get_height()/2, str(atk), va='center')
plt.gca().set_facecolor('#f7f7f7')
plt.grid(axis='x', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```
